```python
%sql

-- Project: Databricks NYC Taxi Lakehouse Analytics
-- Notebook: 04_sql_analytics
-- Objective: Analyze Gold tables using SQL

SHOW TABLES IN workspace.default;
```

```python
%sql

-- Show only NYC Taxi project tables

SHOW TABLES IN workspace.default LIKE '*taxi*';
```

```python
%sql

-- Daily revenue and trip volume

SELECT
    pickup_date,
    total_trips,
    total_revenue,
    avg_total_amount,
    avg_trip_distance,
    avg_trip_duration_minutes
FROM workspace.default.gold_daily_taxi_metrics
ORDER BY pickup_date;
```

```python
%sql

-- Pickup borough ranking by trips and revenue

SELECT
    pickup_borough,
    total_trips,
    total_revenue,
    avg_total_amount,
    avg_fare_amount,
    avg_trip_distance,
    avg_trip_duration_minutes
FROM workspace.default.gold_pickup_borough_metrics
ORDER BY total_trips DESC;
```

```python
%sql

-- Hourly demand and revenue analysis

SELECT
    pickup_hour,
    total_trips,
    total_revenue,
    avg_total_amount,
    avg_trip_distance,
    avg_trip_duration_minutes
FROM workspace.default.gold_hourly_taxi_metrics
ORDER BY pickup_hour;
```

```python
%sql

-- Top 20 pickup zones by trip volume

SELECT
    pickup_borough,
    pickup_zone,
    total_trips,
    total_revenue,
    avg_total_amount,
    avg_fare_amount,
    avg_trip_distance,
    avg_trip_duration_minutes
FROM workspace.default.gold_pickup_zone_metrics
ORDER BY total_trips DESC
LIMIT 20;
```

```python
%sql

-- Executive summary KPIs for January 2024

SELECT
    SUM(total_trips) AS total_trips_january,
    ROUND(SUM(total_revenue), 2) AS total_revenue_january,
    ROUND(AVG(avg_total_amount), 2) AS avg_ticket_daily,
    ROUND(AVG(avg_trip_distance), 2) AS avg_trip_distance_daily,
    ROUND(AVG(avg_trip_duration_minutes), 2) AS avg_trip_duration_daily
FROM workspace.default.gold_daily_taxi_metrics;
```
